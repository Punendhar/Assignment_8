    1. Select 10 records from the adult sqladb

```python
import pandas as pd
import sqlite3
```

```python
adultDf = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data')
ColumnNames= ['age','workclass','fnlwgt','education','education-num','marital-status','occupation','relationship','race','sex','capital-gain','capital-loss','hours-per-week','native-country','(>50K,<=50K)']
adultDf.columns = ColumnNames
```

```python
conn = sqlite3.connect("sqladb.db")
adultDf.to_sql("Adults",conn, if_exists="replace")
```

```python
print(pd.read_sql_query("SELECT * FROM Adults LIMIT 10;", conn))

conn.close()
```

    2. Show me the average hours per week of all men who are working in private sector

```python
import re


adultDf.columns = [re.sub("[- ]", "", col) for col in adultDf.columns]
 

conn = sqlite3.connect("sqladb.db")
adultDf.to_sql("Adults",conn, if_exists="replace")

# To Show Average Hours per week for people working in private sector 

print("\n Average hours per week by men working in private working class are :\n",'-'*80)

sqlQuery = " SELECT round(avg(hoursperweek),2) as  AvgHoursPerWeek  , workclass FROM Adults GROUP BY workclass HAVING workclass LIKE '%Private%' AND sex like '%Male%';"


print(pd.read_sql_query(sqlQuery, conn))
conn.close()
```

    3. To Show frequency Table for Education , Occupation and relationship separately 

```python
adultDf.columns = [re.sub("[- ]", "", col) for col in adultDf.columns]

conn = sqlite3.connect("sqladb.db")
adultDf.to_sql("Adults",conn, if_exists="replace")

print('-'*80,"\nThe Frequency of people based on the Education is :\n",'-'*80)

sqlQuery = " SELECT COUNT(education)  as Frequency , education FROM Adults GROUP BY education  ;"

print(pd.read_sql_query(sqlQuery, conn))

# To Show frequency Table based on Ocupation

print('-'*80,"\nThe Frequency of people based on the Occupation is :\n",'-'*80)
sqlQuery = " SELECT COUNT(occupation)  as Frequency , occupation FROM Adults GROUP BY occupation  ;"
#print(sqlQuery)

print(pd.read_sql_query(sqlQuery, conn))


# To Show frequency Table based on Relationship

print('-'*80,"\nThe Frequency of people based on the RelationShip is :\n",'-'*80)
sqlQuery = " SELECT COUNT(relationship)  as Frequency , relationship FROM Adults GROUP BY relationship  ;"
#print(sqlQuery)

print(pd.read_sql_query(sqlQuery, conn))

conn.close()
```

    4. Are there any people who are married, working in private sector and having a masters degree

```python
adultDf.columns = [re.sub("[- ]", "", col) for col in adultDf.columns]
conn = sqlite3.connect("sqladb.db")
adultDf.to_sql("Adults",conn, if_exists="replace")
sqlQuery = " SELECT Count(*) FROM Adults WHERE education like '%Masters%' AND workclass like '%Private%' AND maritalstatus in "
sqlQuery += " (select distinct maritalstatus from Adults WHERE maritalstatus like '%Married-%' ) ;"
#print(sqlQuery)
print('-'*80,"\n Following nos. of Married people having masters degree and working in private sector :\n",'-'*80)
print(pd.read_sql_query(sqlQuery, conn))
conn.close()
```

    5. What is the average, minimum and maximum age group for people working in different sectors

```python
adultDf.columns = [re.sub("[- ]", "", col) for col in adultDf.columns]
conn = sqlite3.connect("sqladb.db")
adultDf.to_sql("Adults",conn, if_exists="replace")
print('-'*80,"\n Following is minimum , average , maximum age of people working for different sectors :\n",'-'*80)
sqlQuery = " SELECT workclass , min(age), avg(age), max(age) FROM Adults GROUP BY workclass ;"
print(pd.read_sql_query(sqlQuery, conn))
conn.close()
```

    6. Calculate age distribution by country

```python
adultDf.columns = [re.sub("[- ]", "", col) for col in adultDf.columns]
conn = sqlite3.connect("sqladb.db")
adultDf.to_sql("Adults",conn, if_exists="replace")
print('-'*80,"\n Following shows the age distribution by each country :\n",'-'*80)
sqlQuery = " SELECT nativecountry , age , count(*) FROM Adults GROUP BY age ,nativecountry ORDER BY nativecountry ;"
print(pd.read_sql_query(sqlQuery, conn))
conn.close()
```
